## 4. Avaliação dos cenários

```python
results = [evaluate_scenario(scenario) for scenario in scenarios]
[(item['record']['scenario_id'], item['readiness']['decision']) for item in results]
```

## 5. Análise detalhada por cenário

```python
for result in results:
    print(result['record']['scenario_id'])
    print('Decisão:', result['readiness']['decision'])
    print('Energia restante após decolagem:', round(result['energy']['remaining_after_launch_kwh'], 2), 'kWh')
    print('Classificação IA:', result['ai']['classification'])
    print('Anomalias:', result['ai']['anomalies'])
    print('-' * 60)
```
